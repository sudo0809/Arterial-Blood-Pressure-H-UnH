```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# %matplotlib inline
```

```python
class1 = pd.read_excel("final_class1_data.xlsx", index_col='Id')
class1
```

```python
class1['H/UnH'] = np.ones(shape=class1.shape[0])
```

```python
class2 = pd.read_excel("final_class2_data.xlsx", index_col='Id')
class2['H/UnH'] = np.zeros(shape=class2.shape[0])
class2
```

```python
print(class1.isna().any()) 
print()
print(class2.isna().any())
```

```python
class2['Kick Ratio Mean'].fillna(0, inplace=True)
assert class2['Kick Ratio Mean'].isna().any() == False 
```

```python
data = pd.concat([class1, class2])
```

```python
display(data.head())
print(data.shape)
```

```python
data = data.sample(frac=1)
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
data.head()
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x='Sys_mean', y='Dia_mean', data=data, hue='H/UnH')
```

```python
plt.figure(figsize=(12,9))
sns.heatmap(data.corr(), cmap='GnBu', annot=True)
plt.tight_layout()
```

```python
#Scalling the data
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data.iloc[:,:-1], y=data.iloc[:,-1])
scaled_data = pd.DataFrame(data=scaled_data, columns=data.columns[:-1])
```

```python
scaled_data[data.columns[-1]] = data.iloc[:,-1]
scaled_data
```

```python
## Breaking data into Test and Train data
SPLIT = 0.85
m = scaled_data.shape[0]
train_m = int(m*SPLIT)
Train = scaled_data[0:train_m]
Test = scaled_data[train_m:]
```

```python
Train.shape, Test.shape
```

```python
X_train = Train.iloc[:, :-1]
y_train = Train.iloc[:, -1]
X_test = Test.iloc[:, :-1]
y_test = Test.iloc[:, -1]
```

```python
import sklearn
from sklearn.svm import SVC
```

```python
svm = SVC(verbose=1, random_state=28)
```

```python
svm.fit(X_train, y_train)
```

```python
svm.classes_
```

```python
pred_y = svm.predict(X_test)
```

```python
pred_y
```

```python
print(list(y_test))
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, pred_y)
```
